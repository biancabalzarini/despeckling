```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Cargar el archivo CSV
# Reemplazá 'tu_archivo.csv' por el nombre real
df = pd.read_csv("config_simetrico-3conv_256encod.csv")

# Graficar
plt.figure(figsize=(4, 2.5))
plt.plot(df["epoch"][3:], df["loss"][3:], linestyle="-", label="Loss")
plt.xlabel("Época")
plt.ylabel("Error (MSE)")
#plt.title("Evolución del error durante el entrenamiento")
#plt.legend()
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x:.4g}"))
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
